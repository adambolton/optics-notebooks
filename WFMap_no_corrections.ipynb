This notebook uses simulated wavefront errors with no corrections from either
M1 or M2 to characterize the
photometric stability of the resulting telescope PSF over time with respect
to a range of angular field-of-view apertures.

The goal is to understand attainable spectrophotometric accuracy as a function of IFU FOV
size in the worst-case scenario of no mirror control,
and to compare to predictions for ideal and for partially degraded mirror control.

This notebook assumes 1.7 micron wavelength since that is where the PSF will be largest.

```python
# Imports
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import poppy
from poppy.poppy_core import PlaneType
#from poppy.sub_sampled_optics import Subapertures, ShackHartmannWavefrontSensor
import astropy.units as u
import astropy.io.fits as fits
from astropy.table import Table
import copy
import time
import skimage.measure
```

```python
def calc_ensq(im):
    # Simple function to compute ensquared energy for square image
    # im = square input image
    ny, nx = im.shape
    if (ny != nx):
        print("Image must be square")
        return(0)
# 1D image baseline for "what is the smallest square I am in?"
    sqbase = abs(2*np.arange(nx) - nx + 1) + 1
# 2D version of same question (a "bounding-square" coordinate image):
    x_im = np.outer(sqbase, np.full(nx, 1))
    sq_coord = np.maximum(x_im, x_im.T)
# Sort the flattened bounding-square image,
# apply that sort to the data image as well,
# and compute cumulative sum:
    srt = np.argsort(sq_coord.ravel())
    fullbase = sq_coord.ravel()[srt]
    fulldata = im.ravel()[srt]
    fullsum = fulldata.cumsum()
# Downselect to sums fully within squares:
    u_base, u_idx, u_num = np.unique(fullbase, return_index=True, return_counts=True)
    u_idx_2 = u_idx + u_num - 1
    sub_sum = fullsum[u_idx_2]
    return(u_base, sub_sum)
```

```python
# Get simulation data:
xdata_n = fits.getdata('wfe_uncorrected_all.fits',0)
xhdr_n = fits.getheader('wfe_uncorrected_all.fits',0)
n_wfe = xhdr_n['NAXIS3']
# Convert to OPD and transmission:
opd_n = np.nan_to_num(xdata_n)
trans_n = 1. * np.isfinite(xdata_n)
# Create a list of WFE optical elements:
wf_pixscale = xhdr_n['PIXSCL'] * u.m / u.pixel
wfopt_n = []
for idx in range(n_wfe):
    wf_this = poppy.ArrayOpticalElement(transmission=trans_n[idx], opd=opd_n[idx]*1.e-9,
                                        name='Uncorr. WF #'+str(idx+1), pixelscale=wf_pixscale)
    wfopt_n.append(wf_this)
```

```python
# Create error-free wavefront optic for comparison:
wf_perf = poppy.ArrayOpticalElement(transmission=trans_n[0], opd=0.*opd_n[0]*1.e-9,
                                    name='Perfect WF', pixelscale=wf_pixscale)
```

```python
# Define reference parameters:
lambda_ref = 1.7 * u.micron
d_prim = 6.5 * u.m
pxscale = 0.010 # arcseconds per pixel
fov_arcsec = 5.0
```

```python
# Compute PSF for perfect mirror:
osys = poppy.OpticalSystem(pupil_diameter=d_prim)
osys.add_pupil(wf_perf)
osys.add_detector(pixelscale=pxscale, fov_arcsec=fov_arcsec, oversample=1)
plt.figure(figsize=(12,6))
wf_res = osys.calc_psf(lambda_ref, display_intermediates=True)
```

```python
# Compare PSF for randomly chosen wavefront error:
osys = poppy.OpticalSystem(pupil_diameter=d_prim)
osys.add_pupil(wfopt_n[1043])
osys.add_detector(pixelscale=pxscale, fov_arcsec=fov_arcsec, oversample=1)
plt.figure(figsize=(12,6))
wf_res = osys.calc_psf(lambda_ref, display_intermediates=True)
```

```python
# Specify square apertures in arcseconds for which to record ensquared energies:
# (here, out to 4.0 arcsec by 0.2 arcsec.)
ap_base = 0.2 * (np.arange(20.) + 1.)
```

```python
# Loop over the series of wavefronts to compute ensquared PSF energies:
ensq_series = np.full((n_wfe, len(ap_base)), 0.)
for idx in range(n_wfe):
    if (idx % 100 == 0): print(idx)
    osys = poppy.OpticalSystem(pupil_diameter=d_prim)
    osys.add_pupil(wfopt_n[idx])
    osys.add_detector(pixelscale=pxscale, fov_arcsec=fov_arcsec, oversample=1)
    wf_res = osys.calc_psf(lambda_ref, display_intermediates=False, return_intermediates=False)
    u_base, sub_sum = calc_ensq(wf_res[0].data)
    arcsec_base = pxscale * u_base
    ensq_series[idx] = np.interp(ap_base, arcsec_base, sub_sum)
```

```python
# Plot time series of ensquared energies for all apertures:
tbase = xhdr_n['AX3SCL'] * np.arange(n_wfe)
nxplot = 4
nyplot = 5
fig, axis = plt.subplots(nyplot, nxplot)
fig.subplots_adjust(wspace=.4, hspace=.6)
aptitletext = 'aper. = {aper:3.1f} arcsec'
for k in range(len(ap_base)):
    i = k % nxplot
    j = k // nxplot
    ax = axis[j,i].plot(tbase/3600., ensq_series[:,k])
    axis[j,i].set_title(aptitletext.format(aper=ap_base[k]))
    axis[j,i].set_xlabel('hours')
    #axis[j,i].set_ylabel('ensquared energy')

fig.set_figwidth(12)
fig.set_figheight(15)
```

```python
# Compute average, stdev, and absolute variation of ensquared energies in apertures:
avg_ensq = ensq_series.mean(axis=0)
std_ensq = ensq_series.std(axis=0)
min_ensq = ensq_series.min(axis=0)
abd_ensq = np.abs(ensq_series.max(axis=0) - ensq_series.min(axis=0))
rel_ensq = std_ensq / avg_ensq
rel_ensq_ab = abd_ensq / avg_ensq
```

```python
# Punch-line plot:
# Fractional deviation in ensquared energy over time vs aperture size:
plt.figure(figsize=(10,6))
plt.plot(ap_base, np.log10(rel_ensq), label='stdev / mean', lw=2)
plt.plot(ap_base, np.log10(rel_ensq_ab), label='absdev / mean', lw=2)
plt.title('Fractional deviation in ensquared energy over time, no mirror corrections ($\lambda$ = ' +
          lambda_ref.to_string(format='latex') + ')')
plt.xlabel('square aperture size (arcsec)')
plt.ylabel('log10 (dev / mean)')
plt.xlim((0., 4.))
plt.ylim((-5.0,-0.5))
plt.yticks(np.linspace(-5.0, -0.5, 10))
plt.xticks(np.arange(5))
plt.xticks(ap_base, minor=True)
plt.grid(linestyle=':', which='minor')
plt.grid(linestyle='-', which='major', lw=1.5)
plt.legend()
```

```python
res_table = Table()
res_table['aper. (arcsec)'] = ap_base
res_table['aper. (arcsec)'].info.format = '3.1f'
res_table['mean ensq.'] = avg_ensq
res_table['mean ensq.'].info.format = '8.6f'
res_table['stdev ensq.'] = std_ensq
res_table['stdev ensq.'].info.format = '10.8f'
res_table['stdev/mean'] = rel_ensq
res_table['stdev/mean'].info.format = '10.8f'
res_table['absdev ensq.'] = abd_ensq
res_table['absdev ensq.'].info.format = '10.8f'
res_table['absdev/mean'] = rel_ensq_ab
res_table['absdev/mean'].info.format = '10.8f'
#res_table.show_in_notebook()
print(res_table)
```
